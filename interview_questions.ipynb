```python
# Cash Register
# Programming challenge description:
# The goal of this challenge is to design a cash register program. 
# You will be given two decimal numbers. The first is the purchase price (PP) of the item. 
# The second is the cash (CH) given by the customer. 
# Your register currently has the following bills/coins within it:
# 'PENNY': .01,
# 'NICKEL': .05,
# 'DIME': .10,
# 'QUARTER': .25,
# 'HALF DOLLAR': .50,
# 'ONE': 1.00,
# 'TWO': 2.00,
# 'FIVE': 5.00,
# 'TEN': 10.00,
# 'TWENTY': 20.00,
# 'FIFTY': 50.00,
# 'ONE HUNDRED': 100.00

# The aim of the program is to calculate the change that has to be returned to the customer.
# Input:
# Your program should read lines of text from standard input. 
# Each line contains two numbers which are separated by a semicolon.
# The first is the Purchase price (PP) and the second is the cash(CH) given by the customer.

# Output:
# For each line of input print a single line to standard output which is the change to be returned to customer. 
# In case the CH < PP, print out ERROR. 
# If CH == PP, print out ZERO. 
# For all other cases print the amount that needs to be returned, in terms of the currency values provided. 
# The output should be alphabetically sorted.

# Test 1
# Test Input
# Download Test 1 Input
# 15.94;16.00
# Expected Output
# Download Test 1 Output
# NICKEL,PENNY

# Test 2
# Test Input
# Download Test 2 Input
# 17;16
# Expected Output
# Download Test 2 Output
# ERROR

# Test 3
# Test Input
# Download Test 3 Input
# 35;35
# Expected Output
# Download Test 3 Output
# ZERO

# Test 4
# Test Input
# Download Test 4 Input
# 45;50
# Expected Output
# Download Test 4 Output
# FIVE





import sys
# import numpy as np
# import pandas as pd
# from sklearn import ...

register = {
    'PENNY': .01,
    'NICKEL': .05,
    'DIME': .10,
    'QUARTER': .25,
    'HALF DOLLAR': .50,
    'ONE': 1.00,
    'TWO': 2.00,
    'FIVE': 5.00,
    'TEN': 10.00,
    'TWENTY': 20.00,
    'FIFTY': 50.00,
    'ONE HUNDRED': 100.00
}

for line in sys.stdin:
    pp = float(line.split(";")[0])
    ch = float(line.split(";")[1])
    
    # Scenario where Purchase Price is greater than Cash
    if pp > ch:
        print("ERROR")
        break
        
    # Scenario where Purchase Price is exactly equal to the Cash
    if pp == ch:
        print("ZERO")
        break
        
    balance = ch - pp
    
    bill_keys = list(register.keys())
    bill_values = list(register.values())
    
    output_bills_values = []
    output_bills_keys = []
    
    if balance in bill_values:
        print(bill_keys[bill_values.index(balance)])
    
    else:
        for bill in bill_values:
            if balance > bill:
                output_bills_values.append(bill)
        
        output_bills_values = sorted(output_bills_values, reverse = True)
        
        for bills in output_bills_values:
            key_names = bill_keys[bill_values.index(bills)]
            output_bills_keys.append(key_names)
        
        print(",".join(output_bills_keys))
```
